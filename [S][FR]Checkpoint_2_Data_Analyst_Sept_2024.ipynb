### Description des colonnes du DataFrame

 **genre** : Catégorie ou type musical de la piste (par exemple, "Movie").  
 **artist_name** : Nom de l'artiste ou du groupe ayant créé la piste.  
 **track_name** : Titre de la piste musicale.  
 **track_id** : Identifiant unique de la piste.  
 **popularity** : Score indiquant la popularité du morceau (0 à 100).  
 **acousticness** : Niveau de caractère acoustique de la piste (valeur entre 0 et 1).  
 **danceability** : Mesure de la capacité à danser sur la piste (valeur entre 0 et 1).  
 **duration_ms** : Durée totale de la piste en millisecondes.  
 **energy** : Niveau perçu d'énergie du morceau (valeur entre 0 et 1).  
 **instrumentalness** : Proportion de la piste qui est instrumentale (valeur entre 0 et 1).  
 **key** : Tonalité musicale principale de la piste.  
 **liveness** : Probabilité que la piste ait été enregistrée en live (valeur entre 0 et 1).  
 **loudness** : Volume sonore global du morceau (en décibels).  
 **mode** : Mode musical de la piste (Majeur ou Mineur).  
 **speechiness** : Mesure de la présence de paroles dans la piste (valeur entre 0 et 1).  
 **tempo** : Vitesse ou rythme du morceau (battements par minute).  
 **time_signature** : Nombre de temps par mesure (indication rythmique).  
 **valence** : Mesure du caractère positif ou heureux du morceau (valeur entre 0 et 1).


 Train Score :  
Indique la performance du modèle sur l'entraînement. Une valeur élevée montre que le modèle s'ajuste bien aux données.

Mesure la proportion de prédictions correctes. Un score élevé indique une bonne performance globale.

 Precision :  
Mesure la précision des prédictions positives. Un score élevé signifie peu de faux positifs.

 Recall :  
Indique la capacité du modèle à identifier les vrais positifs. Un score élevé signifie moins de faux négatifs.

 F1 Score :  
C'est une moyenne entre précision et rappel. Un F1 élevé signifie un bon équilibre.

  k  :  
Représente le nombre de voisins dans KNN. Un  k  trop bas peut underfit
 un k  élevé peut overfit.
